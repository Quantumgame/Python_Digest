## Работа с файлами 

Для работы с файлами в рython существует функция **open()**:

+ **[Python docs 2.7.11 Input and Output](https://docs.python.org/2/tutorial/inputoutput.html)**

> f = open(file_name, mode)

Как и любая команда, которая требует поиска местонахождения файла, open() ищет файл с именем file_name в той же папке, что и текущий главный модуль. Если файл расположен в другой папке, необходимо указать путь к файлу в его имени (абсолютный или относительный путь).

Параметр mode, который определяет тип файла (текстовый или бинарный, на запись или на чтение) может принимать следующие значения:

+ **r** - только чтение;
+ **w** - только запись;
+ **r+, w+** - и чтение и запись;
+ **a** - дозапись в конец файла. Если файл не существует, то он будет создан;
+ **b** - бинарный формат.
+ **t** - текстовый формат по умолчанию.

Значения могут комбинироваться и объединяться без пробелов: rb+, bw, a.

**N.B.** Чтение и запись информации в текстовый формат через open() **осуществляется с помощью строк**! Т.е. при чтении на выходе будут строки и при записи в файл также необходимо готовить строки!

Основные методы файлов:

+ **read()** - чтение данных из файла. Если параметр не указан, то считывается содержимое всего файла от текущей позиции указателя до конца файла;
+ **readline()** - чтение одной строки файла (до символа перевода строки \n);
+ **readlines()** - чтение данных всего файла в список строк;
+ **write()** - запись одной строки (для текстового формата);
+ **writelines()** - запись строк из последовательности в файл.
+ **close()** - закрытие файла. Следует всегда закрывать файл после того, как он не будет использоваться на чтение и запись.

### Особые символы

При использовании функции open() без использования особых символов все строки будут записываться последовательно в один ряд без пробелов. Обычно текст разбит на абзацы. Для форматирования внешнего вида текствого файла служать специальные символы, среди которых важное место играет символ конца строки (в обычном режиме просмотра текстовых документов его не видно).

В python есть сочетания символов, которые имеют особое значение для интерпретатора. Главным образом они связаны с символом **обратного слэша (обратная косая черта - \)**.

+ \n - перевод строки;
+ \r - возврат каретки;
+ \t - горизонтальная табуляция;
+ \v - вертикальная табуляция.

Если обратный слэш находится в строке рядом с другими символами, то он воспринимается как соответсвующий символ. Однако рекомендуется его "экранировать", т.е. удваивать для избежания недоразумений.

## Методы строк

Как и почти любой другой тип данных, строки имеют ряд особых методов:

+ **lower()** - приведение всех символов строки к нижнему регистру;
+ **upper()** - приведение всех символов строки к верхнему регистру;
+ **isalpha()** - возвращает True, если все символы строки являются буквами;
+ **isdigit()** - возвращает True, если все символы строки строки являются числами;
+ **isspace()** - возвращает True, если все символы строки строки являются пробелами;
+ **startswith('other')** - возвращает True, если строка начинается с подстроки other;
+ **endswith('other')** - возвращает True, если строка заканчивается на подстроку other;

> ЛИСТИНГ 7

Методы поиска в строке:

+ **find('other')** - возвращает индекс первого полного вхождения подстроки other в исходную строку. Возвращает -1, если other полностью не совпадает ни с одной подстрокой оригинальной строки.
+ **index('other')** - возвращает индекс первого полного вхождения подстроки other в исходную строку. Иначе возвращает исключение ValueError.

Методы замены подстрок в строке:

+ **count('other')** - возвращает число полных вхождений (без скользящих пересечений) подстроки other в исходную строку;
+ **replace('old', 'new')** - возвращает строку, где все подстроки 'old' заменяются на подстроки 'new';
+ **strip()** - возвращает строку без пробелов в начале и конце исходной строки;
+ **join(list)** - возвращает строку, полученную из соединения элементов списка с помощью разделителя заданного в качестве строки. Например, '--'.join(['aa', 'bb', 'cc']) -> aa--bb--cc

Методы для превращения строки в список:

+ **split('sep')** - возвращает список, состоящий из подстрок, отделённых разделителем sep. Например, 'aaa,bbb,ccc'.split(',') -> ['aaa', 'bbb', 'ccc']. По умолчанию в s.split() в качестве разделителя используется пробел;

Методы для превращения строки в список:

+ **encode('format')** - возвращает представление строки в выбраном формате format;

## Форматирование строк

Строки могут иметь форматированное представление, т.е. соответствовать некоторым требованиям по расположению и представлению различных типов данных (действительных чисел в первую очередь).

Способы форматирования строк:

+ методы строк **center(), rjust(), ljust()**;

+ связка **'строка' % ()**;

+ метод строк **format()**;

### Простое форматирование

Одним из самых простых методов форматирования строк является выравнивание строк с помощью соответствюущих методов:

+ center(length, 'sym') - по центру;
+ rjust(length, 'sym') - по правому краю;
+ ljust(length, 'sym') - по левому краю.

Можно задать ширину строки length и символ заполнения sym (по умолчанию - пробелы). Таким образом можно привести строки в радующий глаз вид.

> ЛИСТИНГ 10

### Старый способ форматирования строк %

Однако выравнивание не даёт возможность именять представление внутри строки. 

Старый способ форматирования (до версии python 2.6) позволяет представить другие типы данных в виде составной части исходной строки, если добавить в исходную строку особые символы формата, которые начинаются с символа процента %. 

Формат состоит из двух частей соединённых **символом процента %**: левой строки со спец. символами формата и правой, состоящей из переменных.

При добавлении в левую часть спец. символа, начинающегося с символа %, необходимо добавить в правую часть переменную с соответствующим типом данных.

+ %d - целые числа;
+ %f - действительные числа;
+ %s - строки;
+ %g - компактное представление всех типов данных.

Помимо типов данных символ-формат можно добавить длиной, числом значащих цифр после запятой и другими параметрами. 

> Пример '%5d %10f.3 %7s' % (inum, ifloat, istring)

Также форматирование строк позволяет соединять строки более безопасно, чем в случае операции конкатенации (\+).

### Новый способ форматирования строк - метод format()

Начиная с python версии 2.6 форматирование строк можно осуществлять с помощью метода format().

Оно очень похоже на форматирование с помощью знака процента, но имеет ряд преимуществ, а именно позволяет перечисять переменные в произвольном порядке благодаря поддержке именованных и нумерованных указателей. Для быстрой подстановки других типов данных такой способ очень удобен, так как можно опустить описание типа данных и указать лишь один место, где значение той или иной переменнаой должно быть вставлено.

**'Test string {}. Please, {}, come in!'.format(12, 'Peter')**

Вместо спецсимвола формата "%xx" метод format() использует фигурные скобки {}. В самом простом случае кроме них можно ничего не указывать, тогда присваивание правой части (аргументы метода format) будет идти позиционно, по порядку перечисления. 

Помимо позиционного присваивания в {} можно указать номер индекса аргумента, как если бы это был элемент кортежа. Тогда присваивание будет вестись по номерам: 'This is {1}, the capital of {0}'.format('Russia', 'Moscow')

Также можно вместо номеров использовать именованные аргументы в правой части и вставлять такие переменные в строку через имена, заключенные в {}: 'My name is {your_name}'.format(your_name='Pavel')

Такж возможно использование флагов для тонкой настройки вида представления чисел и выравнивания текста.
