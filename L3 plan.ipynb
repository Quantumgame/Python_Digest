## Списки и кортежи

### Методы списков

+ **i** - индекс;

+ **elem** - элемент;

+ **seq** - последовательность.

Методы добавления элементов в список:

+ **append(elem)** - добавляет элемент в конец списка;
+ **extend(seq)** - добавляет элементы последовательности в конец списка;
+ **insert(i, elem)** - добавляет элемент по указанному индексу. Остальные элементы смещаются;

Методы удаления элементов из списка:

+ **pop(i)** - удаляет элемент по указанному индексу и возвращает его;
+ **remove(elem)** - удаляет первый элемент, содержащий указанное значение;

Методы поиска элементов в списке:

+ **count(elem)** - подсчитывает число встречающихся;
+ **index(elem)** - возвращает индекс указанного элемента;

Методы перестановки списка:

+ **reverse()** - обращает список;
+ **sort()** - сортирует список.

Списки можно складывать друг с другом, умножать на простые числа.

### Копирование списков

+ **is** - оператор, который проверяет ссылаются ли переменные на один объект или нет.

Так как при присвоении (оператор "=") в python сохраняется не сам объект (число, строка, список и т.д.), а лишь ссылка на него, то для изменяемых типов данных (списки и словари) возникает проблема копирования.

Проверить, ссылаются ли две ли две переменные на один реальный объект в ОЗУ ПК можно с помощью **оператора is** (ключевое слово).

> Листинг 8

Во время взятия среза происходит т.н. **поверхностное копирование (shallow copy)**. Для простых списков это решает проблему при перенаправлении ссылки в другую переменную. Но т.к. списки могут иметь неограниченную степень вложенности, то для **полного (deep copy)** копирования данных необходимо создавать новый объект используя метод **deepcopy** модуля **copy**.

**N.B.** Т.к. множества и словари не поддерживают срезы, то для их поверхностного копирования используется метод copy(). Для полного или глубокого копирования используется метод **deepcopy** модуля **copy**.

#### Изменение типов последовательностей

Существует несколько способов преобразовывать последовательности друг в друга. 

*A - последовательность.*

+ **list(A)** - преобразование в список;
+ **tuple(A)** - преобразование в кортеж;
+ **set(A)** - преобразование во множество (для словарей - множество ключей);
+ **dict(A)** - преобразование в словарь (требуется два элемента в каждом элементе).

Для создания пустой последовательности можно использовать следующие способы:

+ **[]** - пустой список;
+ **()** - пустой кортеж;
+ **{}** - пустой словарь. Но {1, 2, 3} - множество!

Создать пустое множество скобками нельзя, только с помощью функции set(A). Добавить новое значение в множество можно с помощью метода **add()**.

## Функции перебора без циклов (map и zip)

f - функция, a - последовательность

+ **map(f, a)** - функция, которая позволяет применить функцию f к элементам последовательности a. Возвращает новый список;

В качестве функции можно передавать:

1. пользовательские функции (foo, myfunc);

2. встроенные идентификаторы (**sum**, *int*, **bool**);

3. методы стандратных типов данных (**str.isdigit**, *float.is_integer*).

Функция передаётся без завершающих скобок! Главные условия для функции - она должна поддерживать итерации, а для последовательности - поддерживать итерации!

+ **zip(a1, a2, a3...)** - функция, которая объединяет переданные ей последовательности в список, элементами которого являются кортежи. Возвращает новый список. 

Если количество элементов в переданных последовательностях будет разным, то в zip-список попадут только те элементы, которые существуют во всех последовательностях на одинаковых индексах.

## Работа с модулем random

Модуль random позволяет генерировать последовательности псевдослучайных чисел и осуществлять случайный выбор элементов из последовательностей, а также перемешивать элементы списков случайным образом.

A - последовательность, N - целое число.

+ **random.choice(A)** - возвращает случайный элемент из любой последовательности (строки, списка, кортежа);
+ **random.random()** - возвращает псевдослучайное число в интервале [0;1);
+ **random.sample(A, N)** - возвращает N случайно выбранных элементов из последовательности A;
+ **random.shuffle(A)** - "перемешивает" элементы СПИСКА случайным образом. Функция перемешивает исходный СПИСОК;
+ **random.randint(a, b)** - возвращает случайное ЦЕЛОЕ число из интервала [a, b].

## Алгоритмы сортировки

+ [О сортировках на русском языке](http://programarcadegames.com/index.php?lang=ru);

+ [Анимация некоторых алгоритмов сортировки](http://www.sorting-algorithms.com/).

Алгоритмы сортировки [оцениваются по скорости выполнения и эффективности использования памяти](https://ru.wikipedia.org/wiki/%D0%90%D0%BB%D0%B3%D0%BE%D1%80%D0%B8%D1%82%D0%BC_%D1%81%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B8):

+ **Время** — основной параметр, характеризующий быстродействие алгоритма. Называется также **вычислительной сложностью**. Для упорядочения важны худшее, среднее и лучшее поведение алгоритма в терминах мощности входного множества A. Если на вход алгоритму подаётся множество A, то обозначим n = |A|. Для типичного алгоритма хорошее поведение — это O(n log n) и плохое поведение — это O(n2). Идеальное поведение для упорядочения — O(n);

+ **Память** — ряд алгоритмов требует выделения дополнительной памяти под временное хранение данных. Как правило, эти алгоритмы требуют O(log n) памяти. При оценке не учитывается место, которое занимает исходный массив и независящие от входной последовательности затраты, например, на хранение кода программы (так как всё это потребляет O(1)). Алгоритмы сортировки, не потребляющие дополнительной памяти, относят к сортировкам на месте.

Рассмотрим следующие алгоритмы сортировки:

+ выбором (selection sort);
+ вставками (insertion sort);
+ "пузырьком" (bubble sort).

### Прогоны для сортировки

Чтобы сократить перебор с N\*N, который является худшим вариантом с точки зрения числа итераций, до меньшего числа, можно на каждом шаге внешнего цикла (i) во вложенном цикле рассматривать не всю выборку, а лишь её часть или срез - подвыборку. 

В алгортимах сортировки часто используется приём, когда какой-то участок выборки (это может быть начало, конец выборки или другие части) выбирается в качестве хранилища отсортированных значений. В таком случае, во вложенном цикле нет смысла заново перебирать все элементы выборки, а лишь те, что не были ещё отсортированы.

### Сортировка выбором (Selection sort)

Сортировка выбором является одной из простейших в алгоритмическом плане. В выборке ищется экстремум (мин/макс). Найденный экстремум и первый, либо последний элементы (зависит от направления сортировки и выбранного экстремума) последовательности переставляются. Процедура повторяется для подвыборки, куда не входят переставленные экстремумы.

![caption](files/pics/Selection-Sort-Animation.gif "Рисунок 1 Сортировка выбором")

Сортировка выбором (англ. Selection sort) — поиск наименьшего или наибольшего элемента 
и помещение его в начало или конец упорядоченного списка. Сложность алгоритма: O(n^2).

Описание алгоритма:

1) найти наименьший элемент в массиве
2) поменять местами его и первый элемент в массиве
3) найти следующий наименьший элемент в массиве 
   (исключая найденный экстремум) и поменять местами его и второй элемент массива
4) продолжать это пока весь массив не будет отсортирован

**P.S.** Можно заменить поиск минимума на поиск максимума и вставлять 
найденный экстремум в конец, а не в начало списка.

### Сортировка вставками (Insertion sort)

Сортировка вставками - алгоритм сортировки, в котором элементы входной последовательности просматриваются по одному, и каждый новый поступивший элемент размещается в подходящее место среди ранее упорядоченных элементов[1]. Вычислительная сложность - O(n^2).
