+ арифметических операций:
    + "+, -, * ,/, //, %".

    + ">, <, ==, !=".
+ циклов:
    + for;
    + while.
+ условных операторов:
    + if-elif-else.

    + f = open(filename);

    + f.read;
+ разных последовательностей:
    + списки [];
    + словари {};
    + массивы из модуля numpy;
+ подключение и использование модулей:
    + модулей стандартной библиотеки (os, sys, time, random);
    + сторонних модулей (numpy, pandas).
+ использование функций:
    + def.
+ комментариев:
    + """doc""";
    + # doc.
